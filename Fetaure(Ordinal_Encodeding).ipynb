```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
num_rows = 10

# Generate 10 unique ages (or a range)
ages = np.arange(20, 30) # Example: 20, 21, ..., 29

# Possible values for categorical columns
genders = ['Male', 'Female', 'Non-binary']
reviews = ['Excellent', 'Average', 'Poor']
educations = ['High School', 'Bachelor', 'Master', 'PhD', 'Associate']
purchased_options = ['Yes', 'No']

# Generate random data for other columns ensuring 10 entries
gender_data = np.random.choice(genders, num_rows, replace=True)
review_data = np.random.choice(reviews, num_rows, replace=True)
education_data = np.random.choice(educations, num_rows, replace=True)
purchased_data = np.random.choice(purchased_options, num_rows, replace=True)

# Create the DataFrame
data = {
    'age': ages,
    'gender': gender_data,
    'review': review_data,
    'education': education_data,
    'purchased': purchased_data
}

df = pd.DataFrame(data)
print(df)
```

```python
df.sample(5)
```

```python
df = df[['review','education','purchased']]
df
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['Poor','Average','Excellent'],['High School','Bachelor','Master','PhD','Associate']])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

##OneHotEncoding Using Pand

```python
pd.get_dummies(df,columns=['Fuel Type','Transmission'],drop_first=True)
```

```python
print("""
After applying One-Hot Encoding:

Fuel Type column is converted into:
- Fuel_Type_Electric
- Fuel_Type_Hybrid
- Fuel_Type_Petrol

Transmission column is converted into:
- Transmission_Manual (Automatic is dropped due to drop_first=True)

Each column now contains True/False (or 1/0),
which machine learning models can understand.
""")
```

```python
# Convert boolean to integer (True=1, False=0)
df = df.astype({col: 'int' for col in df.select_dtypes(include='bool').columns})

df.head()
```

```python
# Drop non-numeric columns (for ML preparation)
df_clean = df.drop(['Brand', 'Model'], axis=1)

df_clean.head()
```

```python
print("Final Dataset Info:\n")
df_clean.info()
```

```python
# Average price by condition
avg_price = df.groupby('Condition')['Price'].mean()

print("Average Price by Condition:\n", avg_price)
```

```python
# Select only numeric columns
df_numeric = df_clean.select_dtypes(include=['int64', 'float64'])

# Now apply correlation
corr = df_numeric.corr()

print(corr['Price'].sort_values(ascending=False))
```

```python
print("""
Conclusion:

1. Applied One-Hot Encoding on Fuel Type and Transmission
2. Converted boolean values into numeric format
3. Removed unnecessary text columns
4. Prepared dataset for Machine Learning
5. Performed basic analysis on Price

Now the dataset is ready for model training.
""")
```
