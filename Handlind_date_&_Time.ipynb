```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/orders.csv',usecols = ['order_date','delivery_date'] )
```

```python
df
```

```python
df['Time'] = pd.to_datetime(
    np.random.randint(3600, 86400, len(df)),
    unit='s'
).time
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['order_date'] = df['order_date'].astype('datetime64[ns]')
df['delivery_date'] = df['delivery_date'].astype('datetime64[ns]')
df['Time'] = pd.to_datetime(df['Time'].astype(str))
```

```python
df.info()
```

```python
# extract Year
df['year'] = (df['order_date']).dt.year
df.head()
```

```python
# Extract Month
df['order_month'] = (df['order_date']).dt.month_name()
df['delivary_month'] = (df['delivery_date']).dt.month_name()
df.head()
```

```python
# Extract Day

df['order_day'] = (df['order_date']).dt.day
df['delivary_day'] = (df['delivery_date']).dt.day
df.head()
```

```python
# Which day of week when ordered and delivered
df['order_week_day'] = (df['order_date']).dt.dayofweek
df['delivary_week_day'] = (df['delivery_date']).dt.dayofweek
df.head()
```

```python
df['date_is_weekend_o'] = np.where(df['order_week_day'].isin([6,7]),1,0)
df['date_is_weekend_d'] = np.where(df['delivary_week_day'].isin([6,7]),1,0)
df.head(5)
```

```python
# Extract Week of the Day
df['date_week'] = df['order_date'].dt.isocalendar().week
df['date_week_d'] = df['delivery_date'].dt.isocalendar().week

df.head()
```

```python
# Extract Quater
df['quater'] = df['order_date'].dt.quarter
df['quater_d'] = df['delivery_date'].dt.quarter
df.head()
```

```python
# Extract Semester

df['semster_o'] = np.where(df['quater'].isin([1,2]),1,2)
df['semster_d'] = np.where(df['quater_d'].isin([1,2]),1,2)
df.head()
```

```python
# Difference between order and delivary

df['diff'] = df['delivery_date'] - df['order_date']
df.head()
```

```python
# Month Passed
df['diff_month'] = (
    (df['delivery_date'].dt.year - df['order_date'].dt.year) * 12
    +
    (df['delivery_date'].dt.month - df['order_date'].dt.month)
)

df.head()
```
