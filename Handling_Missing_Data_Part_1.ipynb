```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/data_science_job.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().mean() * 100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape,new_df.shape
```

```python
new_df.hist(bins = 50,figsize = (20,15))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# Original Data
df['training_hours'].hist(bins = 50,ax = ax,color = 'red')
ax.set_title('Original Data')

# New Data
new_df['training_hours'].hist(bins = 50,ax = ax,color = 'yellow')
ax.set_title('Original Data')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# Original Data
df['training_hours'].plot(color = 'black')
ax.set_title('Original Data')

# New Data
new_df['training_hours'].plot(color = 'yellow')
ax.set_title('Original Data')
```
