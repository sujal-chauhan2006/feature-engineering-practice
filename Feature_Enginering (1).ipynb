```python
# Import libraries
import pandas as pd
import numpy as np
```

## Adding simple Dataframe for practise

```python
# Create sample dataset
df = pd.DataFrame({
    'Age': [20, 25, np.nan, 30],
    'Salary': [20000, 30000, 25000, np.nan],
    'Gender': ['Male', 'Female', 'Female', 'Male']
})

print("Original Data:")
print(df)
```

## 1. Handling Missing Values

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

print("\nAfter Handling Missing Values:")
print(df)
```

## 2. Encoding Categorical Data

```python
df_encoded = pd.get_dummies(df, columns=['Gender'])

print("\nAfter Encoding:")
print(df_encoded)
```

## 3. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = df_encoded.copy()

df_scaled[['Age', 'Salary']] = scaler.fit_transform(df_scaled[['Age', 'Salary']])

print("\nAfter Scaling:")
print(df_scaled)
```


## 4. Feature Transformation


```python


# Log transformation (handle skewness)
df_transformed = df.copy()

df_transformed['Salary_log'] = np.log(df_transformed['Salary'])

print("\nAfter Log Transformation:")
print(df_transformed)
```
